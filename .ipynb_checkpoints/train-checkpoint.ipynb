```python
import numpy as np
import pandas as pd
from importlib import reload
import edgeNet as en
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
```

# Data prepration

```python
reload(en)
img_folder = 'data/train/'
data_info = en.data_info(img_folder)
data_info
```

# DataSet and DataLoader

```python
from torchvision import transforms, utils
reload(en)
totensor = en.ToTensor()
data = en.data(img_folder = img_folder, transforms = transforms.Compose([totensor]))
data.summary()
```

* dataLoader

```python
train_loader = DataLoader(data, batch_size=64, shuffle=True, num_workers=0)
```

* check DataLoader

```python
for batch_i, batch_sample in enumerate(train_loader):
    print(batch_i, batch_sample['img'].size(), batch_sample['tar'].size())
    break
```

# Training and Testing

```python
reload(en)
cfg = [[1, 64,64], [64,64,64], [64, 64, 8]]
bks = en.make_blocks(cfg, batch_norm=True)
net = en.net(bks)
net
```

```python
import torch.optim as optim
import torch.nn as nn
import edgeNet as en
import torch.optim as optim
import time
from torch.optim.lr_scheduler import StepLR
reload(en)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.0001, momentum=0.9)
# criterion = nn.NLLLoss()
device='cuda:0'
net.to(device)
log = {'train_loss':[], 'val_loss':[]}
for epoch in range(100):
    t1 = time.time()
    train_loss = en.train(net, device, train_loader, optimizer, 1, criterion)
    val_loss, keep_training= en.val(net, device, train_loader, criterion)
    log['train_loss'].append(train_loss)

    log['val_loss'].append(val_loss)
    
    print('epoch ', epoch + 1, 'loss = [', train_loss,', ', val_loss, ']', time.time()-t1 )
    
    if keep_training==False:
        break
    
```

```python
plt.title('loss')
plt.plot(range(len(log['val_loss'])), log['val_loss'], label='val')
plt.legend()
```
